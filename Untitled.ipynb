```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data4=pd.read_csv("globalterrorismdb_0718dist.csv", low_memory=False)
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

data4.tail(5)
```

```python
plt.figure(figsize=(20,5))
iyear_analysis=sns.countplot(data4["iyear"])
iyear_analysis.set_xticklabels(iyear_analysis.get_xticklabels(),rotation=45, fontsize="x-large")
plt.title("Number of events occured from 1970 to 2017")
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=data4["country_txt"].value_counts()[:20].index, y=data4["country_txt"].value_counts()[:20].values)
plt.xticks(rotation=45, fontsize="x-large")
plt.xlabel("Countries",fontsize="x-large" )
plt.ylabel("Attacks", fontsize="x-large")
```

```python
#Checking for the regions affected the most by terrorism

plt.figure(figsize=(20,10))
sns.countplot(data4["region_txt"])
plt.xticks(rotation=90, fontsize="x-large")
```

```python
plt.figure(figsize=(20,5))
sns.countplot(data4["attacktype1_txt"])
plt.xticks(rotation=70,fontsize="x-large")
```

```python
plt.figure(figsize=(20,10))
sns.countplot(data4["targtype1_txt"])
plt.xticks(rotation=90, fontsize="x-large")
plt.xlabel("Target Type", fontsize="x-large")
plt.ylabel("Attacks", fontsize="x-large")
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=data4["gname"].value_counts()[:20].index, y=data4["gname"].value_counts()[:20].values)
plt.xticks(rotation=90, fontsize="x-large")
plt.xlabel("Terrorist group name", fontsize="x-large")
plt.ylabel("Attacks", fontsize="x-large")
```
